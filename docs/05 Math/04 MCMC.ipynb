```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.linspace(-10, 20, 1000)

def func(x_):
    return 0.3 * np.exp(-2.0 * x_**2) + 0.7 * np.exp(-0.2 * (x_ -10)**2)

y = func(x)

plt.figure()
plt.plot(x, y, "b-")
plt.show()
```

```python
x0 = 7
mcmc_sigma = 2

mcmc_y = stats.norm.pdf(x, x0, mcmc_sigma)

plt.figure()
plt.plot(x, y, "b-")
plt.plot(x, mcmc_y, "r--")
plt.show()
```

```python
recomands = stats.norm.rvs(x0, mcmc_sigma, 2)

recomands
func(x0), func(recomands), x0, recomands
```
